```python
def impute_by_group(X_train, X_test):
    use_mean = ['Lattitude', 'Distance', 'Longtitude', 'Price', 'Landsize', 'BuildingArea']
    use_median = ['Bathroom', 'Bedroom2', 'Car', 'YearBuilt', 'Postcode']
    means = X_train.groupby('Regionname').mean()[use_mean]
    medians = X_train.groupby('Regionname').median()[use_median]
    impute = pd.concat(objs=[means, medians], axis=1)

    train_len = len(X_train)
    datasets = pd.concat(objs=[X_train, X_test], axis=0)
    
    for row in impute.index:
        for col in impute.columns:
            region = datasets.Regionname == row
            datasets[col][region] = datasets[col][region].fillna(impute.loc[row,col])
    
    X_train, X_test = datasets[:train_len], datasets[train_len:]
    return X_train, X_test

def get_dummies(X_train, X_test):
    train_len = len(X_train)
    datasets = pd.concat(objs=[X_train, X_test], axis=0)
    dummies = pd.get_dummies(datasets)
    X_train, X_test = dummies[:train_len], dummies[train_len:]
    
    return X_train, X_test

def preprocess(X, y):
    X_train, X_test = impute_by_group(X_train, X_test)
    X_train, X_test = get_dummies(X_train, X_test)
    return X_train, X_test, y_train, y_test

def run_model(X_train, X_test, y_train, y_test, model):
    model.fit(X_train, y_train)
    cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()
    start = time.time()
    test_score = model.score(X_test, y_test)    
    end = time.time()
    model_time = end - start
    
    return [test_score, cv_score, model_time]
```
