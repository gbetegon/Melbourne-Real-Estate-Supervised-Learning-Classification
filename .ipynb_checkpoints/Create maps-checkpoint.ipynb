```python
import pandas as pd
from bokeh.io import export_png, show
from bokeh.models import GMapOptions
from bokeh.transform import factor_cmap
from bokeh.palettes import brewer, Spectral6
from bokeh.layouts import gridplot
from bokeh.plotting import gmap, figure

path = '/Users/Stephanie/desktop/thinkful/capstone - supervised learning/'
file_name = 'Melbourne_housing_FULL.csv'

df_map = pd.read_csv(path + file_name)
df_map.dropna(inplace=True)

lat_mean = df_map.Lattitude.mean()
lng_mean = df_map.Longtitude.mean()


# Source = https://snazzymaps.com/style/102/clean-grey
style = '''[
    {"featureType": "administrative", "elementType": "labels", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "administrative.country", "elementType": "geometry.stroke", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "administrative.province", "elementType": "geometry.stroke", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "landscape", "elementType": "geometry", "stylers": [
            {"visibility": "on"},{"color": "#e3e3e3"}]},
    {"featureType": "landscape.natural", "elementType": "labels", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "poi", "elementType": "all", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "road", "elementType": "all", "stylers": [
            {"color": "#cccccc"}]},
    {"featureType": "road", "elementType": "labels", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "transit", "elementType": "labels.icon", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "transit.line", "elementType": "geometry", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "transit.line", "elementType": "labels.text", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "transit.station.airport", "elementType": "geometry", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "transit.station.airport", "elementType": "labels", "stylers": [
            {"visibility": "off"}]},
    {"featureType": "water", "elementType": "geometry", "stylers": [
            {"color": "#FFFFFF"}]},
    {"featureType": "water", "elementType": "labels", "stylers": [
            {"visibility": "off"}]}
]'''

size_and_tools = {'plot_height': 600, 'plot_width': 600,
                        'tools':['reset', 'wheel_zoom']}

map_options = GMapOptions(lat=lat_mean, lng=lng_mean, map_type='roadmap', zoom=10, styles=style)

regions = df_map.Regionname.unique()
c_regions = brewer['Set1'][len(regions)]
colormap_regs = dict(zip(regions, c_regions))
c_regions = [colormap_regs[x] for x in df_map.Regionname]

zipcodes = df_map.Postcode.unique()
c_zips = brewer['Set1'][8]
colormap_zips = {}
x=0
for zipcode in zipcodes:
    colormap_zips[zipcode] = c_zips[x]
    x += 1
    if x == 8:
        x = 0
c_zips = [colormap_zips[x] for x in df_map.Postcode]


p1 = gmap('AIzaSyB1Vf_WBp8914WSnBQRxomOOIlZttpmH2Y', map_options, title='Region',
          **size_and_tools)
p2 = gmap("AIzaSyB1Vf_WBp8914WSnBQRxomOOIlZttpmH2Y", map_options, title='Postcode',
         **size_and_tools)

p1.circle(df_map.Longtitude, df_map.Lattitude, size=5, color=c_regions, line_color='black',
          fill_alpha=0.6, line_alpha=0.8, line_width=0.2)

p2.circle(df_map.Longtitude, df_map.Lattitude, size=5, color=c_zips, line_color='black',
         fill_alpha=0.6, line_alpha=0.8, line_width=0.2)

p1.xaxis.major_tick_line_color = None
p1.xaxis.minor_tick_line_color = None

p2.xaxis.major_tick_line_color = None
p2.xaxis.minor_tick_line_color = None


export_png(gridplot([[p1, p2]]), filename='Sales_by_Region_and_Postcode.png')
```
