```python
use_mean = ['Lattitude', 'Distance', 'Longtitude', 'Price', 'Landsize', 'BuildingArea']
use_median = ['Bathroom', 'Bedroom2', 'Car', 'YearBuilt', 'Postcode']

def impute_mean(series):
    return series.fillna(series.mean())
def impute_median(series):
    return series.fillna(series.median())

group_region = df_clean.groupby(['Regionname'])

for col in use_mean:
    df_clean[col] = group_region[col].transform(impute_mean)
    df_someNaN[col] = group_region[col].transform(impute_mean)
for col in use_median:
    df_clean[col] = group_region[col].transform(impute_median)
    df_someNaN[col] = group_region[col].transform(impute_median)
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(df_clean)
```

```python
def compare_models(X, y, model):
    
    start = time.time()
    
    X = pd.get_dummies(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
    
    model.fit(X_train, y_train)
    test_score = model.score(X_test, y_test)
    cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()
    
    end = time.time()
    model_time = end - start
    
    return [test_score, cv_score, model_time]
```

```python
rfc = ensemble.RandomForestClassifier(random_state=13)
lrc = LogisticRegression(penalty='l1')

NaN_imputed = ['All']*3 + ['Some']*3 + ['None']*3
features = ['All', 'Only numerical', 'Only categorical']*3

Xs = []
for dataset in datasets: 
    X_all = dataset.drop(['>3 Rooms'], axis=1)
    X_num = dataset.select_dtypes(include=['float64', 'int']).drop(['>3 Rooms'], axis=1)
    X_cat = dataset.select_dtypes(include=['object'])
    Xs.extend((X_all, X_num, X_cat))

ys = []
for dataset in datasets:
    y = dataset['>3 Rooms']
    ys += [y]*3
```

```python
results = []
for NaN, feat, X, y in list(zip(NaN_imputed, features, Xs, ys)):
    results.append(['R. Forest', NaN, feat, len(y)] + compare_models(X, y, rfc))
    results.append(['L. Regres', NaN, feat, len(y)] + compare_models(X, y, lrc))
```

Hide index

```python
results = pd.DataFrame(results,
                       columns=['Classifier', 'NaN Imputed', 'Features', 'df Length',
                                'CV-Score', 'Test Score', 'RunTime']
                      )
```
