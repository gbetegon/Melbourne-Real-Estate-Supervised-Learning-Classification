## Preprocess data

```python
def impute_by_group(train, test):
    na_cols = train.isna().any()
    use_median = na_cols[na_cols == True].index
    medians = train.groupby('Region').median()[use_median]

    train_len = len(train)
    datasets = pd.concat(objs=[train, test], axis=0)

    for row in medians.index:
        for col in medians.columns:
            region = datasets.Region == row
            datasets[col][region] = datasets[col][region].fillna(medians.loc[row,col])

    train, test = datasets[:train_len], datasets[train_len:]
    return train, test

def get_dummies(X_train, X_test):
    train_len = len(X_train)
    datasets = pd.concat(objs=[X_train, X_test], axis=0)
    dummies = pd.get_dummies(datasets)
    X_train, X_test = dummies[:train_len], dummies[train_len:]
    
    return X_train, X_test

def run_model(X_train, X_test, y_train, y_test, model):
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    cv_score = cross_val_score(model, X_train, y_train, cv=5).mean()
    start = time.time()
    test_score = model.score(X_test, y_test)    
    end = time.time()
    model_time = end - start
    
    return [train_score, cv_score, test_score, model_time]
```

```python
rfc = ensemble.RandomForestClassifier(random_state=13)
lrc = LogisticRegression(penalty='l1', random_state=13)

categoric = list(df_clas.select_dtypes(include=['object']).columns)
features = numeric + categoric
imputing_levels = ['Impute all NaN', 'Impute some NaN', 'Drop all NaN']
feature_selections = ['All features', 'Only numerical', 'Only categorical']

results = []
for level in imputing_levels:
    #split
    train = df_clas[:train_len][features + ['>3 Rooms']] 
    test = df_clas[train_len:][features + ['>3 Rooms']]
    
    # Drop and impute na values as defined
    if level == 'Impute all NaN':
        train, test = impute_by_group(train, test)
        train_all, test_all = train, test # to retrieve later
    elif level == 'Impute some NaN':
        columns = numeric[:]
        columns.remove('BuildingArea')
        train, test = train.dropna(subset=columns), test.dropna(subset=columns)
        train, test = impute_by_group(train, test)
    else:
        train, test = train.dropna(), test.dropna()
    
    for selection in feature_selections:
        if selection == 'All features':
            include = features
        elif selection == 'Only numerical':
            include = numeric
        elif selection == 'Only categorical':
            include = categoric
        
        df_len = len(train) + len(test)
            
        X_train, X_test = train[include], test[include]
        y_train, y_test = train['>3 Rooms'], test['>3 Rooms']
    
        X_train, X_test = get_dummies(X_train, X_test)
        
        for name, model in {'Random Forset': rfc, 'Log. Regression': lrc}.items():
            results.append([name, level, selection, df_len] +
                           run_model(X_train, X_test, y_train, y_test, model)
                          )
```

```python
results = pd.DataFrame(results,
                       columns=['Classifier', 'NaN Imputed', 'Features', 'df Length',
                                'Train Score', 'CV-Score', 'Test Score', 'RunTime']
                      )
```

```python
n_trees = 10
times_from_max = 0

trees = []
scores_train = []
scores_test = []

while times_from_max < 5 and n_trees < 500:
    if not scores_test:
        max_score = 0
    rfc = ensemble.RandomForestClassifier(random_state=13, n_estimators=n_trees)
    rfc.fit(X_train_mix, y_train)
    trees.append(n_trees)
    score_train = cross_val_score(rfc, X_train_mix, y_train, cv=5).mean()
    score_test = rfc.score(X_test_mix, y_test)
    scores_train.append(score_train)
    scores_test.append(score_test)
    if score_test > max_score:
        times_from_max = 0
        max_score = score_test        
    else:
        times_from_max += 1
    n_trees += n_trees
```

```python
plt.plot(trees[:-1], scores_test)
plt.plot(trees[:-1], scores_train)
plt.legend(['Test', 'Cross-Validation'])
```
