## Preprocess data

```python
def get_medians(df_source):
    na_cols = df_source.isna().any()
    use_median = na_cols[na_cols == True].index
    medians = df_source.groupby('Region').median()[use_median]
    return medians

def impute_by_group(df_target, medians):
    for row in medians.index:
        for col in medians.columns:
            region = df_target.Region == row
            df_target[col][region] = df_target[col][region].fillna(medians.loc[row,col])
    return df_target
```

std deviations

```python
rfc = ensemble.RandomForestClassifier(random_state=13)
lrc = LogisticRegression(penalty='l1', random_state=13)

categoric = list(df_clas.select_dtypes(include=['object']).columns)
features = numeric + categoric
imputing_levels = ['Impute all NaN', 'Impute some NaN', 'Drop all NaN']
feature_selections = ['All features', 'Only numerical', 'Only categorical']

results = []
for level in imputing_levels:
    #split
    train = df_clas[:train_len][features + ['>3 Rooms']] 
    
    # Drop and impute na values as defined
    if level == 'Impute all NaN':
        medians = get_medians(train)
        train = impute_by_group(train, medians)
        train_all, medians_all = train, medians # to retrieve later
    elif level == 'Impute some NaN':
        columns = numeric[:]
        columns.remove('BuildingArea')
        train = train.dropna(subset=columns)
        medians = get_medians(train)
        train = impute_by_group(train, medians)
    else:
        train, test = train.dropna(), test.dropna()
    
    for selection in feature_selections:
        if selection == 'All features':
            include = features
        elif selection == 'Only numerical':
            include = numeric
        elif selection == 'Only categorical':
            include = categoric
            
        X_train = train[include]
        y_train = train['>3 Rooms']
    
        X_train = pd.get_dummies(X_train)
        
        for name, model in {'Random Forset': rfc, 'Log. Regression': lrc}.items():  
            scores = cross_validate(model, X_train, y_train, cv=5)
            train_score = scores['train_score'].mean()
            test_score = scores['test_score'].mean()
            confidence = scores['test_score'].std()*2
            fit_time = scores['fit_time'].sum()
            
            results.append([name, level, selection, len(X_train), train_score,
                            test_score, confidence, fit_time])
```

```python
results = pd.DataFrame(results,
                       columns=['Classifier', 'NaN Imputed', 'Features', 'df Length',
                                'Train Score', 'Test Score', '+/- var. 95% CI', 'RunTime']
                      )
```

```python
n_trees = 10
times_from_max = 0

trees = []
test_scores = []
run_times = []

while times_from_max < 4 and n_trees < 600:
    if not test_scores:
        max_score = 0
    rfc_tuned = ensemble.RandomForestClassifier(random_state=13, n_estimators=n_trees)
    scores = cross_validate(rfc_tuned, X_train_mix, y_train, cv=5)
    trees.append(n_trees)
    score = scores['test_score'].mean()
    test_scores.append(score)
    run_times.append(scores['fit_time'].sum())
    if score > max_score:
        times_from_max = 0
        max_score = score        
    else:
        times_from_max += 1
    n_trees += int(n_trees/2)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(trees, test_scores)
plt.ylabel('Score')
 
ax2 = fig.add_subplot(111, sharex=ax1, frameon=False)
ax2.plot(trees, run_times, c='green')
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
plt.ylabel('Run Time')

plt.legend('Score', 'Run Time')

plt.show()
```
